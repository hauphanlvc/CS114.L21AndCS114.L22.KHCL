```python
from bs4 import BeautifulSoup
import lxml
import csv
from urllib.request import Request, urlopen


def get_content(url):
  req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
  webpage = urlopen(req).read()
  # source code : https://stackoverflow.com/questions/16627227/problem-http-error-403-in-python-3-web-scraping
  return webpage
url='https://newsthump.com/'
page = get_content(url)
soup = BeautifulSoup(page,'lxml')
print(soup.prettify())
```

```python
category = soup.find_all("a")
```

# Lấy link dẫn đến các thể loại của trang NewsThump

```python

list_category = category[8:21]
list_url_category = []
for i in list_category:
  list_url_category.append(i.get("href"))
for i in list_url_category:
  print(i)
```

# Lấy dữ liệu từ NewsThump

```python

title = []
href = []
is_sarcastic = []
for i in range(1,1200):
  for path_category in list_url_category:
    
    url = path_category + 'page/' + str(i) + "/"
    try :
      page = get_content(url)
    except:
      pass
    soup = BeautifulSoup(page,'lxml')
    # print(soup.prettify())
    list_news = soup.find_all("h2")
    b = []
    for x in list_news:
      b.extend(x.find_all("a"))
    for x in b:
      if int(x.get("href")[22:26]) <= 2018:
        break 

      is_sarcastic.append(1)
      title.append(x.get("title")[18:])
      href.append(x.get("href"))
    

    
```

```python
import pandas as pd
df = pd.DataFrame()
df['is_sarcastic'] = is_sarcastic
df['headline'] = title
df['article_link'] = href
display(df)
print(len(df))
```

```python
def line(headline,article_link,is_sarcastic):

  line = '{"article_link": ' + '"' + article_link + '", "headline": "' + headline.lower() + '", "is_sarcastic": ' + str(is_sarcastic) + '}\n'
  return line
```

```python
String = []
for i in range(len(df)):
  String.append(line(df['headline'][i],df['article_link'][i],df['is_sarcastic'][i]))
print(String)
```

```python
import json 

with open("NewsThump.json", "w") as outfile:
    for i in range(len(String)):
      dic = json.loads(String[i])      
      json.dump(dic, outfile)
      outfile.write("\n")
```
