# Yêu cầu 2 : Trung bình cộng của Tỷ lệ: problem đã submit/số lượng problem của mỗi assignments ứng với từng user_id 

```python
# Đếm số prolem của từng asignment
dict_prolem_assignment = {} # để lưu trữ các problem của từng assignment
for i in range(len(data_wecode)):
  dict_prolem_assignment[data_wecode["assignment_id"][i]] = {0} 
for i in range(len(data_wecode)):
  dict_prolem_assignment[data_wecode["assignment_id"][i]].add(data_wecode["problem_id"][i])
print(len(dict_prolem_assignment[0]))
dict_prolem_assignment_count = {} # để lưu trữ số lượng problem của từng asignment 
for i in dict_prolem_assignment.keys():
  dict_prolem_assignment_count[i] = len(dict_prolem_assignment[i]) -1 
print(dict_prolem_assignment_count)
```

```python
dict_user_id_assignment_problem = {} # dùng để lưu trữ số lượng prolem của 1 assignment thuộc về 1 user_id
dict_result_user_id_assignment_problem = {} # dùng để tính trung bình cộng của tỷ lệ : prolem đã submit / số lượng của mỗi asignment 
for i in range(len(data_wecode)):
  dict_user_id_assignment_problem[data_wecode["user_id"][i]] = {0}
  dict_result_user_id_assignment_problem[data_wecode["user_id"][i]] = 0
data_wecode_copy = data_wecode.copy()
for x in dict_prolem_assignment.keys():
  for i in range(len(data_wecode_copy)):
    # print(data_wecode_copy["assignment_id"][i])
    if data_wecode_copy["assignment_id"][i] == x :
      dict_user_id_assignment_problem[data_wecode_copy["user_id"][i]].add(data_wecode_copy["problem_id"][i])
      # data_wecode_copy.drop(i,axis = 0)
  # print(len(data_wecode_copy))
  for i in dict_user_id_assignment_problem.keys():
    dict_result_user_id_assignment_problem[i] += (len(dict_user_id_assignment_problem[i]) -1 ) / dict_prolem_assignment_count[x]
for i in dict_user_id_assignment_problem.keys():
    dict_result_user_id_assignment_problem[i] /=len(dict_prolem_assignment_count) 
print(dict_result_user_id_assignment_problem)
```

```python
YeuCau2 = pd.DataFrame(list(dict_result_user_id_assignment_problem.items()),columns=['user_id','Trung Bình Cộng của tỷ lệ '])
print(YeuCau2)
```

# Dự đoán điểm thi cuối kì 

```python
X = Scatter_data[['Sum_PreScore','Trung Bình Cộng của tỷ lệ ']].to_numpy()
Y = Scatter_data['CK'].to_numpy()
```

```python
print(Y) 
```

```python
for i in range(len(Y)):
  try :
    Y[i] = float(Y[i])
  except:
    Y[i] = 0 
print(Y)
```

```python
from sklearn.preprocessing import StandardScaler
X= StandardScaler().fit_transform(X)
Y = Y.reshape(-1,1)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train, Y_test = train_test_split(X,Y,test_size= 0.2)
```

```python
from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics
model = LinearRegression()
model.fit(X_train,Y_train)
pre_y = model.predict(X_test)
mse = metrics.mean_squared_error(Y_test, pre_y)
print("Mean Squared Error {}".format(mse))
```

```python
import math
print("Sai số {}".format((math.sqrt(mse)/Y_test.mean())*100))
```
